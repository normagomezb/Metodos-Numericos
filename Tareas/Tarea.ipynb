```python
import numpy as np
import matplotlib.pyplot as plt

#Defino un metodo para la funcion original
def Funcion(x):
    fun = (1- np.cos(x)/(x**2))
    return(fun)
    
#Evaluar la funcion de 0.1 a 10 aumentando de 0.1 en 0.1 
for i in range(101):
    a = 0.1*i
    Result = Funcion(a)
    print('Para x= {:.1f}       y={}'.format (a, Result))

print()
c=Funcion(1.2e-8)
print('El valor de la funcion en 1.2x10^-8 es: ', c)

#Se crea nuevo modulo para la funcion reescrita 
def New_Funcion(x):
    n_fun = (np.sin(x))**2/( (x**2 * (1 - np.cos(x) )))
    return(n_fun)
#Se evalua nueva funcion    
D = New_Funcion(1.2e-8)
print('La nueva funcion evaluada en el punto 1.2x10^-8 es: ', D)

#Valores para graficar y grafica
valores_x = np.arange(0.1, 10.1, 0.1)
valores_y = Funcion(valores_x)
plt.plot(valores_x, valores_y)
plt.title('')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True)
plt.text(7, -16, r'$f(x) = 1 - \frac{cos(x)}{x^2}$', fontsize=12)
plt.show()
```
