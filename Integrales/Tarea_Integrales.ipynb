```python
#Regla del rectangulo
def Rectangulo(funcion, a, b, N):
    sumatoria = 0
    h = (b-a)/N
    for i in range(1, N):
        xi = a + i*h
        sumatoria += funcion(xi)
    I = h*sumatoria
    return I 

# Regla del Punto Medio
def PuntoM (funcion, a, b, N):
    h = (b-a)/N
    sumatoria = 0
    for i in range(1, N):
        xih = a + i*h + (h/2)
        sumatoria += funcion(xih)
        I = h*sumatoria  
    return I 

#Regla del trapesio: Cuando tendencia de datos es lineal, buena
def Trapecio (funcion, a, b, N):
    Fa = funcion(a)
    Fb = funcion(b)
    h = (b-a)/N
    sumatoria = 0
    for i in range(1, N):
        xi = a + i*h
        sumatoria += funcion(xi)
    I = h*(Fa + 2*sumatoria + Fb )/2
    return I

#Regla de Simpson (1/3)
def Simpson13 (funcion, a, b, N):
    Fa = funcion(a)
    Fb = funcion(b)
    h = (b-a)/N          #N = pares de paneles
    sumatoria1 = 0
    sumatoria2 = 0
    for i in range(1, N):  
        if i%2 != 0:
            xi = a + i*h
            sumatoria1 += funcion(xi)
        else:
            xi = a + i*h
            sumatoria2 += funcion(xi)
    I = h*(Fa + 4*sumatoria1 + 2*sumatoria2 + Fb)/3
    return I

#Regla de Simpson (3/8)
def Simpson38 (funcion, a, b, N):
    Fa = funcion(a)
    Fb = funcion(b)
    h = (b-a)/N
    sumatoria1 = 0
    sumatoria2 = 0
    sumatoria3 = 0
    for i in range(1, N):
        xi = a + i * h
        if i % 3 == 0:
           sumatoria3 += funcion(xi)
        elif i % 3 == 1:
            sumatoria1 += funcion(xi)
        else:
            sumatoria2 += funcion(xi)
    I = (3 * h / 8) * (Fa + 3*sumatoria1 + 3*sumatoria2 + 2*sumatoria3 + Fb)
    return I

#Regla de boole
def Boole(funcion, a, b, N):
    Fa = funcion(a)
    Fb = funcion(b)
    h = (b - a) / N  
    sumatoria1 = 0
    sumatoria2 = 0
    for i in range(1, N):
        xi = a + i * h
        if i % 2 == 0:
            sumatoria1 += funcion(xi)
        else:
            sumatoria2 += funcion(xi)
    xhn = a + h*(N - 1)
    I = (2 * h / 45) * (7 * Fa + 32 * sumatoria2 + 12 * sumatoria1 + 14 * funcion(xhn) + 7*Fb)
    return I
```

```python
#a)
Funcion = lambda x: 3*x**2
a = 0
b = 2

resultado = Rectangulo(Funcion, a, b, 1000)
print('Resultado regla del Rectangulo:  ', resultado)

resultado = PuntoM(Funcion, a, b, 1000)
print('Resultado regla del Punto Medio: ', resultado)

resultado = Trapecio(Funcion, a, b, 1000)
print('Resultado regla del Trapecio:    ', resultado) 

resultado = Simpson13(Funcion, a, b, 1000)
print('Resultado regla de Simpson 1/3:  ', resultado) 

resultado = Simpson38(Funcion, a, b, 1000)
print('Resultado regla de Simpson 3/8:  ', resultado)

resultado = Boole(Funcion, a, b, 1000)
print('Resultado regla de Boole:        ', resultado)
```
